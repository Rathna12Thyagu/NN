```python
wei = 0.10*np.random.randn(784,100)
bias = np.zeros((1,100))
wei1 = 0.10*np.random.randn(100,10)
bias1 = np.zeros((1,10))
```

```python
x=d.drop('label',axis=1)
y=d['label']
inp = x.to_numpy()
inp = inp/255
out = y.to_numpy()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(x))
```

```python
out_1=[]
def forward(inputd):
    for i in range(10):
        z=np.dot(inputd[i],wei) + bias
        z=sigmoid(z[0])
        out_1.append(z)
forward(inp)
#print(out_1)
```

```python
out_2=[]
for i in out_1:
    z = np.dot(i,wei1)+bias1
    z=sigmoid(z[0])
    out_2.append(z)print(out_2)
```

```python
s=[]
for i in range(10):
    esum=0
    for j in range(len(out_2[i])):
        if(out[i]==j):
            esum = esum + (1-out_2[i][j])**2
        else:
            esum = esum + (0-out_2[i][j])**2
    s.append(esum)
print(s)
            
```
